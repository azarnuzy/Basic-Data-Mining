```python
""" Nomor 11 """
import seaborn as sns
data_file = pd.read_csv("file/athlete_events.csv")

#cek median tinggi badan male
med_height_m = data_file.loc[data_file.Sex=="M"]["Height"].median()
med_weight_m = data_file.loc[data_file.Sex=="M"]["Weight"].median()
print(med_height_m)
print(med_weight_m)


# masukan nilai median ke data kosong
data_file.loc[data_file.Sex=="M", "Height"] = data_file.loc[(data_file.Sex=="M", "Height")].fillna(med_height_m)
data_file.loc[data_file.Sex=="M", "Weight"] = data_file.loc[(data_file.Sex=="M", "Weight")].fillna(med_weight_m)

# cek hasil
data_file[(data_file.Sex=="M") & (data_file.Height.isnull()) & (data_file.Weight.isnull())]


```

```python
#cek median tinggi badan male
med_height_f = data_file.loc[data_file.Sex=="F"]["Height"].median()
med_weight_f = data_file.loc[data_file.Sex=="F"]["Weight"].median()
print(med_height_f)
print(med_weight_f)

# masukan nilai median ke data kosong
data_file.loc[data_file.Sex=="F", "Height"] = data_file.loc[(data_file.Sex=="F", "Height")].fillna(med_height_f)
data_file.loc[data_file.Sex=="F", "Weight"] = data_file.loc[(data_file.Sex=="F", "Weight")].fillna(med_weight_f)

# cek hasil
data_file[(data_file.Sex=="F") & (data_file.Height.isnull()) & (data_file.Weight.isnull())]

data_file.isnull().sum()
```
